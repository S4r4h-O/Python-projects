# Load the dataframe

```python
import pandas as pd
df = pd.read_csv("wines/wines.csv")
df
```

# How many wines have been given a rating of 100 points?

```python
len(df.loc[df["points"] == 100])
```

# What is the name of the most expensive wine?

```python
df.loc[df["price"] == df["price"].max()]["name"].squeeze()
```

# Calculate a new column where you show the ratings in a scale from 0 to 5

```python
df["rating_scale"] = df["points"] / 20
```

```python
df
```

# Create a price histogram for wines that costs less than 100

```python
df.loc[df["price"] < 100]["price"].hist()
```

# Plot price horizontally against point vertically. Do you think ratings increase by price?

```python
df.plot(x="price", y="points", figsize=(15, 3), kind="scatter")
```
